# **Guía Orientativa ML - Machine Learning Pipeline**

En este notebook encontrarás una guía orientativa paso a paso para llevar a cabo un proyecto de Machine Learning completo. Está pensada para seguirse de forma ordenada, aunque en la práctica algunos pasos son iterativos.

Un pipeline de ML bien construido nos va a permitir pasar de los datos en bruto a un modelo entrenado, evaluado y guardado, con decisiones justificadas en cada paso.

**¿Cuáles son los objetivos que debe cubrir un proyecto de Machine Learning?**
- Resolver un problema de negocio concreto mediante modelos predictivos o de agrupación.
- Explorar y preparar los datos orientando cada decisión al modelado.
- Comparar modelos de forma rigurosa y seleccionar el más adecuado.
- Optimizar el modelo elegido y evaluarlo contra datos no vistos.
- Comunicar los resultados ligándolos siempre con el problema original.

**¿Cuándo necesitamos Machine Learning y no solo un EDA?**
- Cuando queremos **predecir** un valor o categoría para nuevos datos (supervisado).
- Cuando queremos **descubrir grupos o estructuras** en los datos sin etiquetas (no supervisado).
- Cuando el análisis manual no escala y necesitamos automatizar decisiones.
- Como paso natural después de un EDA que ha validado la calidad de los datos y las relaciones entre variables.

### 1. Business Case & Problem Definition

1. **Business understanding**: identifica el problema de negocio que quieres resolver. ¿Qué decisión mejorará si tienes un modelo que funcione bien?

2. **Hipótesis y objetivo de modelado**: define qué quieres predecir o descubrir. Si es supervisado, ¿cuál es la variable target? Si es no supervisado, ¿qué estructura buscas en los datos? Ten claro el **tipo de problema**: clasificación, regresión, clustering, etc.
3. **Plan de acción**: ¿qué harás con el modelo una vez entrenado? ¿Qué decisiones tomará el negocio en función de sus predicciones?
4. **Requerimientos de los datos**: lista los datos necesarios a alto nivel. ¿Cuántos registros necesitas mínimo? ¿Qué variables son imprescindibles?
5. **Disponibilidad**: primer punto de inflexión. Si el dato necesario no existe o no es accesible, hay que replantear el proyecto.
6. **Adquisición de datos**: ve a por los datos — bases de datos, datasets públicos, APIs, Web Scraping.
7. **Calidad**: segundo punto de inflexión. Valida que los datos son suficientes, representativos y no tienen problemas graves de calidad que impidan el modelado.

### 2. Data Understanding

8. **Exploratorio inicial**: realiza un primer análisis de las variables disponibles — número de filas y columnas, naturaleza de las variables, tipos de datos, proporción de missings y primeros registros.

9. **Tabla de variables**: inventaría todas las variables: nombre, tipo de dato, naturaleza (numérica/categórica/fecha), aportación esperada al modelo y observaciones. Identifica ya aquí cuál es tu **variable target**.

10. **División train/test**: **antes de cualquier exploración profunda**, separa el dataset en train y test. El conjunto de test no debe tocarse hasta la evaluación final. Todo el EDA y el preprocesado se realiza sobre train.

### 3. EDA Dirigido al Modelado

A diferencia del EDA clásico, aquí el análisis exploratorio está al servicio del modelo. Cada gráfica o estadístico debe responder a una pregunta orientada al modelado.

11. **Análisis de la variable target**: si es un problema supervisado, analiza en profundidad tu variable objetivo.

    11.1 En regresión: distribución, asimetría, outliers, transformaciones necesarias (log, sqrt...)  
    11.2 En clasificación: balanceo de clases, proporción de cada categoría, necesidad de técnicas de balanceo (oversampling, undersampling, class_weight)

12. **Relación features vs. target**: analiza cómo se relaciona cada variable independiente con el target. Esto te ayudará a identificar las features más predictivas.

    12.1 Numérica vs. target numérico: correlación, scatter plots  
    12.2 Numérica vs. target categórico: boxplots por clase, distribuciones por clase  
    12.3 Categórica vs. target: tablas de contingencia, tasas de target por categoría

13. **Correlaciones entre features**: detecta multicolinealidad entre variables independientes. Features muy correlacionadas pueden perjudicar algunos modelos.

    13.1 Heatmap de correlación  
    13.2 Pairplot de variables seleccionadas  

14. **Detección de outliers**: identifica valores atípicos que puedan distorsionar el entrenamiento.

    14.1 Histogramas y diagramas de caja  
    14.2 Método IQR  
    14.3 Z-score para distribuciones aproximadamente normales

### 4. Preprocesado & Feature Engineering

15. **Unión de tablas**: si los datos vienen de varias fuentes, únelos mediante merge o join antes de preprocesar.

16. **Eliminar columnas irrelevantes**: identificadores únicos, columnas constantes, columnas con excesivos missings (>60%), variables que no aporten información al modelado.

17. **Formato de los datos**: fuerza los tipos correctos en cada columna. Un numérico cargado como string generará errores silenciosos en todo el pipeline.

18. **Tratamiento de missings**: en ML, los missings deben imputarse antes del modelado.

    18.1 Variables numéricas: imputación por media, mediana o con KNNImputer, o eliminación  
    18.2 Variables categóricas: imputación por moda o con una categoría "Desconocido"  
    18.3 **Importante**: los parámetros de imputación se calculan sobre train y se aplican sobre test (fit en train, transform en test)

19. **Tratamiento de outliers**: decide si eliminar, capear (winsorizing) o mantener según el impacto en el modelo.

20. **Duplicados**: elimina registros duplicados.

23. **Feature Engineering**: crea nuevas variables a partir de las existentes que puedan aportar más información al modelo.

    23.1 Ratios entre variables numéricas  
    23.2 Extracción de componentes de fechas (año, mes, día de la semana)  
    23.3 Binning de variables continuas  
    23.4 Interacciones entre features

21. **Encoding de variables categóricas**: los modelos de ML trabajan con números, no con strings.

    21.1 Label Encoding: para variables ordinales (baja, media, alta)  
    21.2 One-Hot Encoding: para variables nominales sin orden  
    21.3 Target Encoding: para variables con alta cardinalidad  
    21.4 **Importante**: fit en train, transform en train y test por separado

22. **Escalado de variables numéricas**: necesario para modelos basados en distancias o en gradiente.

    22.1 StandardScaler: media 0, desviación estándar 1 (recomendado para la mayoría)  
    22.2 MinMaxScaler: escala entre 0 y 1  
    22.3 **No es necesario** para modelos basados en árboles (Random Forest, XGBoost, etc.)  
    22.4 **Importante**: fit en train, transform en train y test por separado

24. **Selección de features**: reduce la dimensionalidad eliminando variables poco relevantes.

    24.1 Feature importance de modelos de árboles  
    24.2 Correlación con el target  
    24.3 RFE (Recursive Feature Elimination)

### 5. Modelado

25. **Definición de la métrica de evaluación**: antes de entrenar ningún modelo, define qué métrica vas a usar para comparar. La elección depende del tipo de problema y del contexto de negocio.

    **Clasificación:** Accuracy, Precision, Recall, F1-Score, ROC-AUC  
    **Regresión:** MAE, MSE, RMSE, R²  
    **Clustering:** Silhouette Score, Davies-Bouldin, Elbow Method  
    
    Justifica siempre la elección: en un modelo médico, el Recall puede ser más importante que el Accuracy.

26. **Modelo baseline**: entrena un modelo simple que sirva como referencia mínima. Si tu modelo avanzado no supera el baseline, algo va mal.

    26.1 Clasificación: DummyClassifier (predict most frequent), LogisticRegression simple  
    26.2 Regresión: DummyRegressor (predict mean), LinearRegression simple  
    26.3 Clustering: KMeans con k=2 o k basado en el número de grupos esperado

27. **Comparativa de modelos**: prueba varios algoritmos con sus parámetros por defecto y compara sus métricas usando validación cruzada.

    27.1 Modelos lineales: LogisticRegression, LinearRegression, Ridge, Lasso  
    27.2 Modelos basados en árboles: DecisionTree, RandomForest, GradientBoosting, XGBoost  
    27.3 Otros: KNN, SVM, Naive Bayes  
    27.4 Documenta la comparativa en una tabla con las métricas de cada modelo

### 6. Optimización de Hiperparámetros

28. **Selección del modelo a optimizar**: elige el modelo o los 2-3 mejores modelos de la comparativa anterior para optimizar sus hiperparámetros.

29. **Estrategia de búsqueda**: elige la técnica de búsqueda de hiperparámetros más adecuada.

    29.1 **GridSearchCV**: exhaustiva, prueba todas las combinaciones. Recomendada cuando el espacio de búsqueda es pequeño  
    29.2 **RandomizedSearchCV**: aleatoria, más eficiente para espacios grandes  
    29.3 Siempre con **validación cruzada** (cv=5 habitualmente) para evitar sobreajuste  
    29.4 Optimización Bayesiana

30. **Validación cruzada**: usa cross_val_score o StratifiedKFold para problemas de clasificación desbalanceada. La validación cruzada solo se aplica sobre el conjunto de **train**.

31. **Análisis de resultados de la búsqueda**: documenta los mejores hiperparámetros encontrados y el impacto en la métrica respecto al modelo con parámetros por defecto.

### 7. Evaluación Final

32. **Evaluación contra test**: con el modelo final optimizado, evalúa por primera (y única) vez sobre el conjunto de **test**. Este resultado es el rendimiento real esperado en producción.

33. **Análisis de resultados**: interpreta las métricas obtenidas.

    33.1 **Clasificación**: matriz de confusión, reporte de clasificación, curva ROC  
    33.2 **Regresión**: distribución de residuos, predicción vs. valor real, análisis de errores  
    33.3 **Clustering**: visualización de clusters, análisis de centroides, Silhouette Score

34. **Interpretabilidad del modelo**: entiende qué variables son más importantes para el modelo.

    34.1 Feature importance para modelos de árboles  
    34.2 Coeficientes para modelos lineales  
    34.3 SHAP values para explicabilidad avanzada (opcional)

35. **Contraste con el problema de negocio**: ¿el rendimiento del modelo es suficiente para resolver el problema planteado? ¿Qué impacto tendría en el negocio? Sé honesto con las limitaciones del modelo.

### 8. Persistencia y Entregables

36. **Guardar el modelo**: exporta el modelo (y el preprocesador si aplica) en formato `pickle` o `joblib` en el directorio `src/models/`.

    36.1 Guarda tanto el modelo como los transformadores (scaler, encoder) que necesites para inferencia  
    36.2 Documenta qué archivos se guardan y cómo cargarlos

37. **Limpieza del notebook**: prepara el `main.ipynb` final.

    37.1 Elimina celdas de prueba y alternativas descartadas  
    37.2 Añade headers claros en cada sección  
    37.3 Verifica que se ejecuta de principio a fin sin errores  
    37.4 Documenta las decisiones técnicas más importantes con comentarios en markdown

38. **Documentación en GitHub**: organiza el repositorio con la estructura requerida, completa el README.md y asegúrate de que el código es reproducible.

39. **Comunicación y presentación de resultados**: prepara la presentación siguiendo la estructura recomendada: problema de negocio → planteamiento técnico → dataset y EDA → modelado y resultados → conclusiones.
