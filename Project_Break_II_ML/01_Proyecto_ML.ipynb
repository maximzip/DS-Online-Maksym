## 1. Objetivo del Proyecto

El objetivo del proyecto de Machine Learning es evaluar vuestras capacidades para construir un modelo de ML que resuelva un problema concreto **de principio a fin**, sin considerar la parte de productización o industrialización.

Para ello, se pide desarrollar un **proceso de ML completo** que dé lugar a una serie de entregables, sobre los que realizar una exposición en vídeo del trabajo y los resultados. La aproximación puede ser **supervisada, no supervisada, o híbrida**.

### Lo que se valorará

**Se valorará la resolución sensata, el seguimiento razonado de los pasos aprendidos, y el análisis claro y bien expuesto de los resultados ligándolos con el problema que se quiere resolver.**

### Elección del problema

Podéis escoger el dataset y problema de negocio que queráis siempre que defendáis con argumentos que este problema tiene algún tipo de solución o ayuda a su solución empleando Machine Learning.

### Restricciones

> **Importante:** NO aceptaremos ningún proyecto realizado sobre los datasets ya resueltos en las Live Reviews.

> **Importante:** No utilicéis datasets "famosos" o de práctica común como Titanic, Iris, Boston Housing o California Housing.

### EDA/miniEDA dirigido

Sobre los datos elegidos será necesario hacer un **EDA o miniEDA dirigido** orientado a entender los datos en función del problema de modelado:

- Análisis de la variable objetivo (si aplica)
- Identificación de las features más relevantes
- Detección de patrones, outliers y relaciones entre variables
- Justificación de las decisiones de preprocesado y feature engineering

## 5. Git Flow para Trabajo en Equipo

Trabajar en equipo sobre el mismo repositorio requiere una estrategia de ramas clara. Seguiréis un **Git Flow simplificado** adaptado a proyectos de ML en equipo de 2 o 3 personas.

### Estructura de ramas

```
main          ← rama de producción, solo código revisado y funcional
  └── develop ← rama de integración, aquí se unen el trabajo de todos
        ├── feature/eda            ← exploración y análisis de datos
        ├── feature/preprocessing  ← limpieza y feature engineering
        ├── feature/modeling       ← entrenamiento y comparativa de modelos
        └── feature/evaluation     ← evaluación final y persistencia
```

### Reglas del equipo

- **Nunca se trabaja directamente sobre `main` ni sobre `develop`.**
- Cada tarea o bloque de trabajo tiene su propia rama `feature/`.
- Antes de hacer merge a `develop`, otro miembro del equipo revisa el código (**Pull Request**).
- Solo se hace merge a `main` cuando el proyecto está completo, revisado y sin errores.
- Los **commits deben ser frecuentes y descriptivos**: un commit por cada avance significativo, no uno al final del día con todo.

### Flujo de trabajo paso a paso

**1. Configuración inicial del repositorio (solo una vez, lo hace un miembro)**
```bash
git clone https://github.com/usuario/ML_proyecto.git
cd ML_proyecto
git checkout -b develop
git push -u origin develop
```

**2. El resto del equipo clona y activa develop**
```bash
git clone https://github.com/usuario/ML_proyecto.git
cd ML_proyecto
git checkout develop
```

**3. Crear rama de feature antes de empezar a trabajar**
```bash
git checkout develop
git pull origin develop        # siempre actualizar antes de crear rama
git checkout -b feature/preprocessing
```

**4. Trabajar y commitear con frecuencia**
```bash
git add .
git commit -m "feat: imputación de missings con mediana en variables numéricas"
git commit -m "feat: one-hot encoding aplicado a variables categóricas"
git commit -m "fix: corregido data leakage en el escalado del test set"
```

**5. Subir la rama y abrir Pull Request hacia develop**
```bash
git push -u origin feature/preprocessing
# GitHub → Pull Requests → New Pull Request
# Base: develop  ←  Compare: feature/preprocessing
```

**6. Otro miembro revisa, aprueba y hace merge**
```bash
# Tras el merge en GitHub, actualizar develop en local
git checkout develop
git pull origin develop
```

**7. Merge final a main al terminar el proyecto**
```bash
git checkout main
git merge develop
git push origin main
git tag -a v1.0 -m "Entrega final Proyecto ML"
git push origin v1.0
```

> ⚠️ **Importante — Tag de versión final obligatorio:** Una vez hecho el merge a `main`, **debéis crear y publicar el tag `v1.0`** como se muestra arriba. El tag es la marca oficial de vuestra entrega: permite a los evaluadores acceder exactamente al estado del repositorio en el momento de la entrega, aunque después sigáis haciendo cambios. Sin el tag, la entrega no se considerará cerrada.

### Convención de mensajes de commit

| Prefijo | Cuándo usarlo |
|---------|---------------|
| `feat:` | Nueva funcionalidad o paso del pipeline |
| `fix:` | Corrección de un error |
| `docs:` | README, comentarios o markdown |
| `refactor:` | Mejora del código sin cambiar su comportamiento |
| `data:` | Adición o modificación de datos |
| `model:` | Entrenamiento, evaluación o guardado de modelos |

### Reparto sugerido para un equipo de 3

| Miembro | Rama | Responsabilidad |
|---------|------|-----------------|
| Persona A | `feature/eda` | EDA dirigido, análisis del target y features |
| Persona B | `feature/preprocessing` | Preprocesado, encoding y escalado |
| Persona C | `feature/modeling` + `feature/evaluation` | Comparativa de modelos, optimización y evaluación final |

> **Consejo:** Aunque cada persona tenga su rama principal, revisad juntos los Pull Requests. La revisión de código es donde más se aprende.
