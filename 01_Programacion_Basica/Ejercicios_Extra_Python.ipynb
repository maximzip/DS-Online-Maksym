## Ejercicio 1
Crea un programa que recoja dos inputs. Tiene que comprobar si su suma es igual, superior o inferior a 10

````python
num1 = int(input("Introduzca el primer numero: "))
num2 = int(input("Introduzca el segundo numero: "))

def numeros_in(num1, num2):
    if num1 + num2 == 10:
        return "La suma es igual a 10"
    elif num1 + num2 < 10:
        return "La suma es inferior a 10"
    elif num1 + num2 > 10:
        return "La suma es superior a 10"
    else:
        return "Error al introducir los numeros"

print(numeros_in(num1, num2))
````
