## Ejercicio 3

Crea una función que permita eliminar los elementos de una lista o tupla, incluso los repetidos. Para ello la función debe admitir un número arbitrario de elementos a eliminar, la lista o tupla que debe eliminarse, el número de elementos repetidos que debe eliminarse, este parámetro por defecto debe estar a None, que quiere decir todos los elementos repetidos.

Si el elemento no está en la lista o tupla debe imprimirlo por pantalla y no dar error.


````python
def eliminar_elementos(secuencia, *elementos, repeticiones=None):
    # Convertimos a lista para poder modificar
    resultado = list(secuencia)

    for elem in elementos:
        if elem not in resultado:
            print(f"El elemento {elem} no está en la secuencia")
            continue

        if repeticiones is None:
            # Elimina todas las apariciones
            while elem in resultado:
                resultado.remove(elem)
        else:
            # Elimina un número concreto de repeticiones
            count = 0
            while elem in resultado and count < repeticiones:
                resultado.remove(elem)
                count += 1

    # Devolver el mismo tipo que la entrada
    if isinstance(secuencia, tuple):
        return tuple(resultado)
    return resultado

lista = [1, 2, 2, 3, 2, 4]

print(eliminar_elementos(lista, 2))
````

## Ejercicio 5

Dividir entre 100 los elementos positivos de una lista

````python
L = [10, 5, -1, 20, 30, -100, 0]

# Método tradicional
resultado = []
for elemento in L:
    if elemento > 0:
        resultado.append(elemento/100)
    else:
        resultado.append(elemento)

print(resultado)
````

````python
L = [10, 5, -1, 20, 30, -100, 0]
# Con comprensión de listas
resultado = [x / 100 if x > 0 else x for x in L]
print(resultado)
````
