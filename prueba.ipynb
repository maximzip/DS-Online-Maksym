```python
import urllib.request
from pathlib import Path

url = "https://github.com/StackExchange/Survey/raw/refs/heads/main/packages/archive/2025/results.csv"
destino = Path("../data/survey_results_2025.csv")
destino.parent.mkdir(parents=True, exist_ok=True)

urllib.request.urlretrieve(url, destino)
print("Descargado correctamente")
```

```python
import pandas as pd

df = pd.read_csv("../data/survey_results_2025.csv")
print(df.shape)
print(df.columns.tolist())
df
```

```python
pd.set_option('display.max_columns', None)
print(df.columns.tolist())
```

```python
for col in df.columns:
    print(f"{col} ({df[col].dtype}): {df[col].dropna().unique()[:3]}")
```

```python
resumen = pd.DataFrame({
    "tipo": df.dtypes,
    "nulos_%": (df.isnull().sum() / len(df) * 100).round(1),
    "únicos": df.nunique(),
    "ejemplo": [df[col].dropna().iloc[0] if df[col].notna().any() else "todo nulo" 
                for col in df.columns]
})
resumen
```

```python
columnas_interes = [
    "ConvertedCompYearly",
    "Country",
    "YearsCode",
    "EdLevel",
    "DevType",
    "OrgSize",
    "RemoteWork",
    "LanguageHaveWorkedWith",
    "AISelect",
    "JobSat"
]

df_eda = df[columnas_interes]
print(df_eda.shape)
df_eda.head(15)
```

```python
# Ver todos los valores únicos de DevType
for valor in df_eda["DevType"].dropna().unique():
    print(valor)
```

```python
# Filtrar filas que contengan cualquier rol de datos
roles_data = [
    "Data scientist",
    "Data analyst",
    "Data engineer",
    "Machine learning"
]

# Construir el patrón de búsqueda
patron = "|".join(roles_data)

df_data = df_eda[df_eda["DevType"].str.contains(patron, case=False, na=False)]

print(f"Filas totales:     {len(df_eda)}")
print(f"Filas data roles:  {len(df_data)}")
print(f"Porcentaje:        {len(df_data)/len(df_eda)*100:.1f}%")
```
